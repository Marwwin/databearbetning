## Vi shoppar på newegg

```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://www.newegg.com/Desktop-Graphics-Cards/SubCategory/ID-48?Tid=7709&PageSize=96")
page = req.text
page
```

```python
soup = BeautifulSoup(page,"html.parser")
```

```python
listElms = soup.findAll('li',{"class":"price-current"})
listElms[0].strong.text
```

```python
prices = []
for elm in listElms:
    if len(elm.findAll('strong')) > 0:
        prices.append(elm.findAll('strong'))
prices.
```
