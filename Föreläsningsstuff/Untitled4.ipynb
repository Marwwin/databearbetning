```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
st = requests.get("https://store.steampowered.com/specials#p=0&tab=NewReleases")
soup = BeautifulSoup(st.text,"html.parser")
```

```python
sales = soup.find_all('a',{"class":"tab_item"})
```

```python
sales[0]
```

```python
flatten = lambda f : [item for sublist in f for item in sublist]
```

```python
games = [[g.findAll("div",{"class":"discount_original_price"}), g.findAll("div",{"class":"tab_item_name"}),g.findAll("span",{"class":"top_tag"}),g.findAll("div",{"class":"discount_pct"}),g.findAll("div",{"class":"discount_final_price"}),g.findAll("span",{"class":"platform_img"})] for g in sales]
```

```python

te = list(map(lambda x: {"Name":x[1][0].text,"Original Price":x[0][0].text,"Discount":x[3][0].text,"Final Price":x[4][0].text,"Genre":list(map(lambda y: y.text,x[2])),"Platform":[p for p in flatten(list(map(lambda t: t["class"],x[5]))) if p != "platform_img"]},games))

#te[5]["Platform"]
df = pd.DataFrame(te)
df
```
