### Beautifulsoup

är en modul som man använder för att parsa html kod.



```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(hd.text,'html.parser')
```

```python
print(soup.find_all('a'))
```

```python
soup.find_all("span")[2]
```

```python
soup.findAll("span")[2]["class"]
```

```python
soup.table["class"]
```

```python
rows = []
for row in soup.findAll("table")[2].findAll("td"):
    if (len(row.text)<=10):
        rows.append(row.text)
        
rows
```

```python
table = soup.findAll("table")[2]
```

```python
# List comprehension med if sats
[row.text for row in table.findAll("td") if (len(row.text) <=5)]
```
